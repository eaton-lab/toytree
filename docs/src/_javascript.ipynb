```python
import xml.etree.ElementTree as xml

import numpy as np
import toyplot
import toyplot.html
from IPython.display import HTML

# 1. Setup Data - Include labels in the data array to avoid the label list error
n_rows, n_cols = 10, 100
raw_data = np.random.choice(["A", "C", "G", "T"], size=(n_rows, n_cols))
header = np.array([[str(i) for i in range(n_cols)]])
# Stack header on top of data
table_data = np.vstack([header, raw_data])

color_map = {"A": "#1f77b4", "C": "#ff7f0e", "G": "#2ca02c", "T": "#d62728"}

# 2. Build the Canvas
width, height, table_width = 400, 350, 3000
canvas = toyplot.Canvas(width=width, height=height)
# We no longer pass 'label=[...]' to avoid the TypeError
table = canvas.table(table_data, bounds=(0, table_width, 20, height - 30))

# 3. Apply Styles
table.cells.cell.lstyle = {"text-anchor": "middle", "alignment-baseline": "middle"}

for i in range(n_rows):
    for j in range(n_cols):
        val = raw_data[i, j]
        # table_data includes headers at index 0, so DNA starts at i+1
        table.cells.cell[i + 1, j].style = {"fill": color_map[val]}
        table.cells.cell[i + 1, j].lstyle = {"fill": "white", "font-weight": "bold"}

# 4. Use the Internal Pipeline
# We call render() without a file object to get the XML root back
# This function internally creates the RenderContext correctly
root_xml = toyplot.html.render(canvas)

# Generate a unique ID for the SVG if it doesn't have one
svg_element = root_xml.find("svg")
canvas_id = svg_element.get("id")

# Create the JS string
js_code = f"""
(function() {{
    const element = document.querySelector("#{canvas_id}");
    let vbX = 0;
    const maxScroll = {table_width} - {width};
    
    element.setAttribute("viewBox", `0 0 {width} {height}`);
    element.style.cursor = "grab";
    element.style.userSelect = "none";
    
    let isDragging = false;
    let startX = 0;
    
    element.addEventListener("mousedown", (e) => {{
        isDragging = true; startX = e.clientX; element.style.cursor = "grabbing";
    }});
    
    element.addEventListener("mousemove", (e) => {{
        if (!isDragging) return;
        const dx = e.clientX - startX;
        vbX = Math.max(0, Math.min(vbX - dx, maxScroll));
        element.setAttribute("viewBox", `${{vbX}} 0 {width} {height}`);
        startX = e.clientX;
    }});
    
    const stop = () => {{ isDragging = false; element.style.cursor = "grab"; }};
    element.addEventListener("mouseup", stop);
    element.addEventListener("mouseleave", stop);
}})();
"""

# Append the JS directly into the HTML fragment inside a script tag
script_xml = xml.SubElement(root_xml, "script", type="text/javascript")
script_xml.text = js_code

# 5. Export and Display
html_fragment = xml.tostring(root_xml, encoding="unicode", method="html")
display(HTML(html_fragment))
```

```python
import numpy as np
import toyplot
import toyplot.html
from IPython.display import HTML, display

# 1. Generate Mock DNA Data (10 rows x 100 columns)
np.random.seed(42)
bases = ["A", "C", "G", "T"]
data = np.random.choice(bases, size=(10, 100))

# 2. Initialize the Canvas and Table
viewport_width = 400
canvas = toyplot.Canvas(width=viewport_width, height=350)
table_width = 3000
table = canvas.table(data, bounds=(0, table_width, 20, 320))

# 3. Apply DNA State Colors (Using the correct .lstyle attribute!)
color_map = {"A": "#1f77b4", "C": "#ff7f0e", "G": "#2ca02c", "T": "#d62728"}

for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        val = data[i, j]
        # Coloring the background block and making text white for a true viewer look
        table.cells.cell[i + 1, j].style = {"fill": color_map[val]}
        table.cells.cell[i + 1, j].lstyle = {
            "fill": "white",
            "font-weight": "bold",
            "text-anchor": "middle",
            "alignment-baseline": "middle",
        }

        # Centering styling for ALL cells (headers and data)
        table.cells.cell.lstyle = {
            "fill": "white",
            "text-anchor": "middle",  # Horizontal center
            "alignment-baseline": "middle",  # Vertical center
        }

# 4. Extract HTML and wrap it in our Div
raw_html = toyplot.html.tostring(canvas)
wrapped_html = f'<div id="dna-viewport-container">{raw_html}</div>'

# 5. Inject Mouse Drag JavaScript
js_string = """
<script>
(function() {
    const container = document.getElementById("dna-viewport-container");
    if (!container) return;
    
    const svg = container.querySelector("svg");
    if (!svg) return;
    
    // Setup ViewBox parameters
    let vbX = 0;
    const vbY = 0;
    const vbWidth = 400;  
    const vbHeight = 350; 
    const maxScroll = 3000 - vbWidth; 
    
    svg.setAttribute("viewBox", `${vbX} ${vbY} ${vbWidth} ${vbHeight}`);
    
    // Make it look draggable
    svg.style.cursor = "grab";
    svg.style.userSelect = "none"; // Prevents text highlighting while dragging
    
    // State variables for dragging
    let isDragging = false;
    let startX = 0;
    
    // 1. Mouse Down: Start dragging
    svg.addEventListener("mousedown", function(e) {
        isDragging = true;
        startX = e.clientX;
        svg.style.cursor = "grabbing";
    });
    
    // 2. Mouse Move: Pan the viewBox
    svg.addEventListener("mousemove", function(e) {
        if (!isDragging) return;
        
        e.preventDefault();
        
        // Calculate how far the mouse moved
        const dx = e.clientX - startX;
        
        // Subtract dx to move the view in the opposite direction of the mouse
        vbX = vbX - dx; 
        
        // Keep the view strictly within the table boundaries
        vbX = Math.max(0, Math.min(vbX, maxScroll));
        
        // Update SVG and reset startX for the next frame
        svg.setAttribute("viewBox", `${vbX} ${vbY} ${vbWidth} ${vbHeight}`);
        startX = e.clientX;
    });
    
    // 3. Mouse Up / Leave: Stop dragging
    function stopDrag() {
        isDragging = false;
        svg.style.cursor = "grab";
    }
    
    svg.addEventListener("mouseup", stopDrag);
    svg.addEventListener("mouseleave", stopDrag); // Stops drag if mouse exits the plot
})();
</script>
<p style="font-family: sans-serif; font-size: 12px; color: #666; margin-top: 5px;">
    <em><strong>Click and drag</strong> horizontally to pan through the sequence.</em>
</p>
"""

# 6. Render in Notebook
display(HTML(wrapped_html + js_string))
```
