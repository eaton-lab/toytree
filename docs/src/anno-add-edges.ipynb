# add edges

```python
import toytree

tree = toytree.rtree.bdtree(ntips=15, seed=123)
```

```python
c, a, m = tree.draw()

tree.annotate.add_edges(a, color="idx", width=8, use_color_gradient=True, opacity=0.5)

toytree.save(c, "/home/deren/Desktop/TEST.svg")
toytree.save(c, "/home/deren/Desktop/TEST.pdf")
toytree.save(c, "/home/deren/Desktop/TEST.html")
```

```python
import toyplot

h = toyplot.html.render(c)
```
