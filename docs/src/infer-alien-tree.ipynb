## Example 1: Gene is transferred and retained at original placement

#### Label the target and destination of the transfer event

```python
stree = toytree.rtree.unittree(50, seed=1234)
stree.set_node_data(
    "color",
    {80: "lightblue", 70: "orange"},
    inplace=True,
    inherit=True,
    default="white",
)
stree = stree.root("r28").unroot()
stree.draw(
    "s",
    layout="d",
    height=300,
    node_sizes=15,
    node_colors="color",
    node_mask=False,
    label="Species tree",
);
```

#### Create a gene tree representing one transfer event

```python
src = stree.get_mrca_node("r32", "r33")
dst = stree.get_mrca_node("r24", "r25")
_tmp = stree.copy()
_tmp = _tmp.set_node_data(
    "name", {i: f"H-{i.name}" for i in _tmp[src.idx].get_leaves()}
)
hgt_tree = stree.mod.add_internal_node_and_subtree(
    dst, subtree=_tmp[src.idx], subtree_rescale=True, parent_name="HGT"
)
hgt_tree.draw(
    "s",
    layout="d",
    height=300,
    node_colors="color",
    node_sizes=15,
    node_mask=False,
    label="gene tree",
);
```

#### Create an IMAP mapping gene tree tip names to species

```python
imap = {}
for node in hgt_tree:
    if node.is_leaf():
        imap[node.name] = node.name.lstrip("H-")
```

### Compute the alien tree index score

```python
result = alien_tree_index(
    stree, hgt_tree, imap=imap, min_candidate_species=1, min_candidate_tips=1
)
```

```python
src = 175
dst = 187
_tmp = dtree.copy()
_tmp = _tmp.set_node_data("name", {i: f"H-{i.name}" for i in _tmp[src].get_leaves()})
dhgt_tree = dtree.mod.add_internal_node_and_subtree(
    dst, subtree=_tmp[src], subtree_rescale=True, parent_name="HGT"
)

src = 171
dst = 130
dhgt_tree.set_node_data(
    "color",
    {src: "cyan", dst: "magenta"},
    inplace=True,
    inherit=True,
)
_tmp = dhgt_tree.copy()
_tmp = _tmp.set_node_data("name", {i: f"H-{i.name}" for i in _tmp[src].get_leaves()})
dhgt_tree = dhgt_tree.mod.add_internal_node_and_subtree(
    dst, subtree=_tmp[src], subtree_rescale=True, parent_name="HGT"
)

imap = {}
for node in dhgt_tree:
    if node.is_leaf():
        imap[node.name] = node.name.lstrip("H-")

dhgt_tree.draw(
    "s",
    layout="d",
    height=300,
    node_colors="color",
    node_sizes=15,
    node_mask=False,
    label="gene tree",
);
```

```python
species = toytree.tree("(((a,b),(c,d)),(e,f));")
gene = toytree.tree("((((ga,gc),gb),gd),(ge,gf));")
imap = {"ga": "a", "gb": "b", "gc": "c", "gd": "d", "ge": "e", "gf": "f"}
```

```python
species.draw("s");
```

```python
gene.draw("s");
```

```python
result = alien_tree_index(species, gene, imap, error_mode="none")
result["summary"]
```

```python
result["data"].sort_values("gain_weighted", ascending=False).head(10)
```

```python
best_idx = result["summary"]["best_node_idx"]
node_colors = ["black"] * gene.nnodes
if best_idx is not None:
    node_colors[best_idx] = "crimson"
gene.draw(layout="d", tip_labels=True, node_sizes=8, node_colors=node_colors);
```

## Example 2: Multi-copy gene family with explicit mapping

```python
gene_multi = toytree.tree("(((((ga1,ga2),gb1),gc1),gd1),(ge1,gf1));")
imap_multi = {
    "ga1": "a",
    "ga2": "a",
    "gb1": "b",
    "gc1": "c",
    "gd1": "d",
    "ge1": "e",
    "gf1": "f",
}
res_multi = alien_tree_index(species, gene_multi, imap_multi, error_mode="local_nni")
res_multi["summary"]
```

```python
res_multi["data"].sort_values("gain_weighted", ascending=False).head(10)
```

```python
best_idx = res_multi["summary"]["best_node_idx"]
node_colors = ["black"] * gene_multi.nnodes
if best_idx is not None:
    node_colors[best_idx] = "crimson"
gene_multi.draw(layout="d", tip_labels=True, node_sizes=8, node_colors=node_colors);
```

## Notes

- `imap` is required and maps each gene tip to a species tip.
- `alien_index` is the maximum weighted RF gain over tested internal nodes.
- `error_mode="local_nni"` adds one-step local topological alternatives around each tested edge.
