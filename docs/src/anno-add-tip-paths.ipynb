# add tip paths

The annotation method `add_tip_paths` adds path (line) elements relative to tips on a tree drawing.

```python
import toytree

tree = toytree.rtree.unittree(10, seed=123)
```

## Examples

Path elements extend outwards from the tips of a tree to a max distance of ``depth`` distance, defined in px units, with a default of 100.

```python
c, a, m = tree.draw(width=400)
m1 = tree.annotate.add_tip_paths(axes=a, depth=100, color="salmon");
```

Paths can be used to represent data or relationships, for example by pointing to different end positions, using different path lengths, or styling paths by color, opacity, tooltip hovers, and curvature.

```python
c, a, m = tree.draw(width=400)

# a list of ordered end spans in range(0, ntips)
spans = [7, 4, 0, 2, 1, 5, 6, 8, 3, 9]

# draw tip paths to end 'spans'
m1 = tree.annotate.add_tip_paths(
    axes=a,
    spans=spans,
    color=("idx", "Spectral"),
    opacity=0.5,
    offset_start=0,
    style={"stroke-width": 4},
    bezier_fractions=(0.45, 0.55),
);
```

## ``ends`` and ``depth``

``ends`` and ``spans`` set the coordinates of the end points of the paths. Depending on the tree drawing orientation/layout, this may coorespond to positions along the x or y axis. As shown in the examples, above, one or both of these can be left at the default ``None`` which activates fitting.

When ``spans`` is None the position of path ends on the span-axis is set the same as their starts, i.e., the lines will be straight. If you set the spans to differnet values then you will get curved or angled path lines.

When ``ends`` is None the position of path ends on the depth-axis will be set by the backup arg ``depth``, which sets the width of the area in front of the tips that will contain the paths in pixel units. If you make ``depth`` larger then the paths will use up more space relative to the tree and tip labels. (Note: increase the ``.draw(width...)`` arg to increase the space for both).

By contrast, if ``ends`` is non-empty, then it sets the end position of the paths in data units, i.e., the depth units of the tree. (See [aligning path](#aligning-paths) for more details on data versus pixel units).

```python
# draw the tree
c, a, m = tree.draw(width=400, scale_bar=True)

# add paths ends in data units (same scale as the tree depth)
m1 = tree.annotate.add_tip_paths(
    axes=a, ends=[1, 2, 3, 1, 2, 3, 3, 2, 1, 2], color="salmon"
);
```

## ``span`` and ``bezier_fraction``

The ``spans`` arg to ``add_tip_paths`` sets the end positions of paths along the span-axis. The span-axis is the axis along which tips are separated. For a 10 tip tree the default span positions are ``range(0, 10)`` (see the [coordinates and axes](drawing-coordinates) for more details). When ``span=None`` the end positions are the same as the start positions, so you get straight lines. A ``span`` arg of ntips values can be entered to change the end positions. 

The curvature of a path from start to end position is controlled by the ``bezier_fraction`` argument. This essentially sets 2 points: the proportion of length the path will run straight before curving towards the end points, and the proportion of length from the end when it will stop curving. The defalut is ``(0.45, 0.55)``. A value of ``(0, 1)`` would draw a straight angled line.

Using these args you can do some fun stuff like drawing curved paths towards different or shared end points, like below.

```python
# draw a tree
c, a, m = tree.draw(width=400)

# add tip path annotations
sinks = [2, 2, 2, 5, 5, 5, 2, 5, 8, 8]
m1 = tree.annotate.add_tip_paths(
    axes=a,
    ends=[1] * tree.ntips,
    spans=sinks,
    style={"stroke": "grey", "stroke-width": 4, "stroke-opacity": 0.3},
    bezier_fractions=(0.45, 0.55),
)

# add text labels assigned to tip coordinates, then displaced in px units in depth.
# [you could do a similar thing using annotation.add_tip_text]
a.text(
    a=[1, 1, 1],
    b=[2, 5, 8],
    text=["community 1", "community 2", "community 3"],
    style={
        "fill": "#262626",
        "text-anchor": "start",
        "-toyplot-anchor-shift": 10,
    },
);
```

## Aligning paths
``ends`` and ``depth`` provide two ways to set the end positions of paths on the depth axis. Use ``ends`` when trying to position paths to align with other Marks plotted in data units (e.g., other trees on the same Cartesian), and use ``dpeth`` when trying to position paths to align with other Annotation Marks that are plotted in pixel units *relative* to tree data.

### data units
We can use the ``ends`` and ``spans`` to draw paths connecting the tip nodes between two trees, or connecting to other data marks such as from a scatterplot or barplot. 

In this example, we draw a second tree on the same axes using `xbaseline` to shift its position by 2.0 data units. 

To align the paths, we can now just set ends to 2.0 for every tip. In the example below, we also use ``offset_start`` and ``offset_end`` to offset the path start and end positions from the tips, to instead align near the tip labels, by shifting their positions in pixel units.

See the [co-phylogeny plot cookbook](cookbook-cophylogeny) for more examples.

```python
# get two trees with differen topologies
tree1 = toytree.rtree.unittree(ntips=10, seed=123)
tree2 = toytree.rtree.unittree(ntips=10, seed=333, random_names=True)

# the distance between the two trees in data (dist) units
DIST = 2.0

# draw the first tree facing right
c, a, m1 = tree1.draw(layout="r", width=500, height=300)

# draw the second tree facing left, displaced along x-axis by DIST data units
_, _, m2 = tree2.draw(axes=a, layout="l", xbaseline=DIST)

# draw paths from tree 1, set 'depth' to find dist to other tips in px units
m3 = tree1.annotate.add_tip_paths(
    axes=a,
    ends=[DIST] * tree1.ntips,
    spans=[tree2.get_nodes(i)[0].idx for i in tree1.iter_tip_labels()],
    color="idx",
    opacity=0.6,
    style={"stroke-width": 4},
    offset_start=30,
    offset_end=-30,
)
```

### px units

In some cases you may want to draw paths that connect to other annotation Marks that are placed relative to the tree by some offset of pixel units. For example, here we add tip markers at 40 and 140 px units from the tips. We can add paths connecting these markers by using the ``offset_start`` arg to shift the start position by 40 pixels. And we can use the ``depth`` arg to set the end position of the paths at 140 pixels. Note that if you enter values for the ``lengths`` (data units) it will override the ``depth`` arg. 


```python
# draw the tree
c, a, m = tree.draw(width=400)

# add markers offset from +40 to +100 px units from the tree
m1 = tree.annotate.add_tip_markers(a, xshift=40, color="salmon", style={"stroke": None})
m2 = tree.annotate.add_tip_markers(
    a, xshift=140, color="salmon", style={"stroke": None}
)

# add paths connect the anno markers using their positions in px units
m3 = tree.annotate.add_tip_paths(
    axes=a,
    spans=[0, 1, 2, 6, 4, 3, 5, 7, 9, 8],
    style={"stroke": "salmon", "stroke-width": 4, "stroke-opacity": 0.5},
    offset_start=40,
    depth=140,
    bezier_fractions=(0.45, 0.55),
);
```

## Styling paths

### Style dict
You can set CSS styles to the path elements using the ``style`` dict arg. 

```python
c, a, m = tree.draw(width=400)
m1 = tree.annotate.add_tip_paths(
    axes=a,
    style={
        "stroke": "salmon",
        "stroke-width": 3,
        "stroke-opacity": 0.5,
        "stroke-dasharray": "4,8",
        "stroke-linecap": "round",
    },
)
```

### Color and Opacity
You can set per-path ``color`` and ``opacity`` entered as a single value; a sequence of ntips values; as a feature name that is stored to the tree; or as a tuple mapping feature data to a colormap or value range. Note the ``color`` and ``opacity`` override similar arguments in the ``style`` dict.

```python
c, a, m = tree.draw(width=400)
m1 = tree.annotate.add_tip_paths(
    axes=a,
    color=("idx", "BlueRed"),
    opacity=[0.5, 0.6, 0.7, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 0.8],
)
```

## Mask
The ``mask`` arg can be used to only show paths extending from a subset of tips. The ``ToyTree.get_tip_mask`` method is helpful for creating a boolean mask to select tip nodes by name or idx label.

```python
c, a, m = tree.draw(width=400)
m1 = tree.annotate.add_tip_paths(
    axes=a,
    mask=tree.get_tip_mask(2, 3, 4, 5),
    color="red",
    offset_start=35,
)
```
