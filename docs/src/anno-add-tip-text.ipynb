# add tip text

The ``add_tip_text`` method is used to add text labels relative to the tip node positions on a tree drawing. It can be used to add tip labels to a tree, providing a greater variety of styling options that the ``tip_labels`` argument to ``ToyTree.draw()``. Or, it can be used to add additional text annotations to a tree drawing to indicate additional information. 

```python
import toytree

NEWICK = """((32082_przewalskii:0.00259326350604,33588_przewalskii:0.00247133969857)100:0.0179371130487,((33413_thamno:0.00565358258838,(30556_thamno:0.00653218253975,((40578_rex:0.00335406134691,35855_rex:0.00339963302434)100:0.00222999650239,(35236_rex:0.00580524693404,(39618_rex:0.000962081181627,38362_rex:0.00109218434613)100:0.00617527349892)96:0.000738900380519)99:0.000783365499905)100:0.00103379657491)100:0.00538723112355,((29154_superba:0.00634236826448,30686_cyathophylla:0.00669944988924)100:0.00237994755604,(41954_cyathophylloides:8.88803116778e-05,41478_cyathophylloides:5.28218433585e-05)100:0.00941020878048)100:0.00297626149201)100:0.0179371130487);"""
tree = toytree.tree(NEWICK)
```

## Example
In this simplest case you can add tip labels to a tree that was originally drawn without them, at which point you can now apply a larger suite of styling options.


```python
# draw a tree w/o tip labels
c, a, m = tree.draw(width=400, height=300, tip_labels=False)

# add tip labels
tree.annotate.add_tip_text(axes=a, labels="name");
```

For example, you can provide a custom set of ``labels``, set their distance ``offset_depth`` from the tip nodes in px units, set the tip label color and opacity, font, angle, and more.

In this example we select the "name" feature from the tree, and use options `italic`, `bold`, and `fn` to transform the displayed labels without having to change the stored names on the tree.

```python
# draw a tree with aligned tip but no labels
c, a, m = tree.draw(width=400, height=325, tip_labels=False, tip_labels_align=True)

# add styled tip labels to the tree
tree.annotate.add_tip_text(
    axes=a,
    labels="name",
    italic=True,
    bold=True,
    font_size=13,
    fn=lambda x: x.split("_")[0],
    offset_depth=15,
    color="salmon",
    opacity=0.8,
);
```

The method is not only useful for tip labels, but for adding any additional text anchored by node tip positions. For example, trait values.

```python
# draw a tree with aligned tip and labels
c, a, m = tree.draw(width=450, height=300, tip_labels_align=True)

# add 'dist' feature as offset monospace text with 6 floating points
tree.annotate.add_tip_text(
    axes=a,
    labels="dist",
    font_family="monospace",
    offset_depth=175,
    color="grey",
    fn=lambda x: f"{float(x):.6f}",
);
```

## Tip positions

When using `add_tip_text` the text is added *relative* to the positions of the tip nodes. The tip nodes are positioned on the depth and span axes as described in [Coordinates and axes](drawing-coordinates), and vary depending on the layout. 

### alignment
Beware that when trees are drawn with `tip_labels_align=True` this effectively aligns the tip node positions, such that annotations added to the tree tips will also have aligned starting positions.

### offset
You can shift the position of added text relative to these positions in units of pixels using `offset_depth` and `offset_span`. 

These arguments have consistent effects whether trees are facing down, left, circular, fan-shaped, etc.

```python
# draw a tree with no labels
c, a, m = tree.draw(height=350, tip_labels=False, layout="d")

# add tip labels
tree.annotate.add_tip_text(axes=a, labels="name", offset_depth=10, italic=True);
```

## Labels
You can set the ``label`` arg to enter either a sequence of text strings of length ntips, or as a str feature name of data stored to the tree. The default is to select the "name" feature.

```python
# draw a tree with aligned tips but no labels
c, a, m = tree.draw(height=350, tip_labels=False)

# add custom labels
tree.annotate.add_tip_text(
    axes=a, labels=range(tree.ntips), offset_depth=10, italic=True
);
```

## Mask
``mask`` is used to subselect which text labels will be shown. It is convenient to use ``ToyTree.get_tip_mask`` to select tips by name or idx.

```python
# draw a tree w/o tip labels
c, a, m = tree.draw(width=450, height=300, tip_labels_align=True)

# add tip labels
tree.annotate.add_tip_text(
    axes=a,
    labels="idx",
    mask=tree.get_tip_mask("~.*(_p|_c|_s).*"),  # select by regex
    # mask=tree.get_tip_mask(0, 1, 9, 10, 11, 12),  # or select by idx
    color="salmon",
    offset_depth=160,
    bold=True,
    text_anchor="middle",
);
```

## Style
### Color and opacity
The color and opacity args can be used to set per-datum styles. These can be entered as a single value, a sequence of ntips values, as a str for a feature that is stored to the tree, or using the tuple-format to map a feature to a colormap. The example below maps the feature "idx" to the Spectral (default) colormap. 

```python
# draw a tree w/o tip labels
c, a, m = tree.draw(width=400, height=300, tip_labels=False, tip_labels_align=True)

# add tip labels
tree.annotate.add_tip_text(axes=a, labels="name", color="idx", opacity=1.0);
```

## Style dict
Most options that can be styled using the style dict are available as primary arguments in ``add_tip_text``.

```python
# draw a tree w/o tip labels
c, a, m = tree.draw(width=400, height=300, tip_labels=False, tip_labels_align=True)

# add tip labels
tree.annotate.add_tip_text(
    axes=a,
    style={
        "fill": "darkcyan",
        "fill-opacity": 0.6,
        "-toyplot-anchor-shift": 10,
        "font-size": 14,
        "text-anchor": "start",
    },
);
```
