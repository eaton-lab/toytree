# Planning host/companion Cartesian updates

Host axes remain the fit-driving axes for tree, annotation, and later host marks.
Tree and annotation scale bars render on separate hidden companion Cartesians, and
each companion keeps interactive coordinate readouts disabled so only the host axes
reports cursor coordinates.

```python
import toytree
```

```python
# tree = toytree.tree("https://eaton-lab.org/data/Cyathophora.tre").root("~prz")
```

```python
# scale bar should extend from 0 to 2 on the x-axis
tree = toytree.rtree.unittree(20, seed=123, treeheight=2)
tree.draw(layout="c", scale_bar=True, width=400);
```

```python
# scale bar should extend from 0 to 80 on the y-axis
tree = toytree.rtree.unittree(20, seed=123, treeheight=100)
tree.draw(layout="c60-320", scale_bar=True, width=400);
```

```python
# scale bar should extend from 0 to past -80 on the x-axis, and use positive labels
tree = toytree.rtree.unittree(20, seed=123, treeheight=100)
tree.draw(layout="c60-160", scale_bar=True, width=400);
```

```python
# scale bar should extend from 0 to 2 on the y-axis
tree = toytree.rtree.unittree(20, seed=123, treeheight=2)
tree.draw(layout="c60-160", scale_bar=True, width=400);
```

```python
tree = toytree.rtree.bdtree(10, seed=123)
c, a, m = tree.draw(
    tip_labels_align=True,
    tip_labels_style={"anchor-shift": 10},
    scale_bar=True,
    padding=90,
    width=400,
    layout="r",
)
tree.annotate.add_tip_bars(
    a,
    "dist",
    color="dist",
    width=0.80,
    style={"stroke-width": 1.5, "stroke": "#262626"},
);
```

```python
tree = toytree.rtree.bdtree(10, seed=123)
c, a, m = tree.draw(
    tip_labels_align=True, scale_bar=True, padding=90, width=600, layout="u"
)
# m1 = tree.annotate.add_tip_bars(a, 'dist', style={"stroke-width": 3, "stroke": "red"}, width=0.99);
m1 = tree.annotate.add_tip_tiles(a, color="idx", offset=0, depth=None)
```

```python
# add tree to host Cartesian A; return A
c, a, m = tree.draw(width=400)

# create companion Cartesian B for the tree scale bar; return A
tree.annotate.add_axes_scale_bar(axes=a)

# create companion Cartesian C for the annotation scale bar; return A
tree.annotate.add_tip_bars(axes=a, values="dist", offset=50, depth=20)

# add data Marks to host Cartesian A
a.scatterplot(range(0, 10), range(10));

# A now keeps fitting the tree Mark, annotation Mark, and later data Mark while B and C stay scale-only.
```

```python
# add tree to host Cartesian A; return A
c, a, m = tree.draw()

# create companion Cartesian B for the tree scale bar; return A
tree.annotate.add_axes_scale_bar(axes=a, padding=10)

# create companion Cartesian C for the annotation scale bar; return A
tree.annotate.add_tip_bars(axes=a, values="dist", offset=50, depth=100);

# A remains the fit-driving host axes for the tree Mark and annotation Mark after B is added.
```

```python
# add tree to host Cartesian A; return A
c, a, m = tree.draw(scale_bar=True)

# create companion Cartesian C for the annotation scale bar; return A
tree.annotate.add_tip_bars(axes=a, values="dist", offset=50, depth=100);

# A still fits the tree Mark and annotation Mark when the tree scale bar was created by draw(scale_bar=True).
```
