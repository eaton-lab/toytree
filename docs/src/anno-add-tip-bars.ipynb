# Add tip bars

The annotation ``add_tip_bars`` adds a bar chart to display a quantitative trait next to tips of a tree.

Bars are normalized to the largest tip value and drawn into the tip slots defined by neighboring tip positions.


```python
import toytree

# an example tree
tree = toytree.rtree.bdtree(ntips=15, seed=123)

# simulate a categorical and quantitative trait
tree.pcm.simulate_continuous_trait(name="Y", root_state=5, inplace=True, seed=321);
```

## Example
The tip bar markers are automatically placed beyond the farthest tip label extent, with values scaled relative to each other to fit within a pixel width of 100px. This makes it easy to view the bars as data and to optionally apply a distinct scale bar to the data.

In this example I select the continuous trait "Y".

```python
# draw a tree
canvas, axes, m0 = tree.draw(layout="r", width=350)

# add bars mapped to tips on the tree
tree.annotate.add_tip_bars(axes=axes, data="Y");
```

## Bar offset, depth, width

``offset`` sets the start position of bars. The default (None) starts from 10px past the farthest tip label. You can override this by setting a numeric value: 0 starts from tips, any positive value shifts the bars away from the tips by that number of pixels.

``depth`` sets size of the region used to represent the bars. Larger values will give more space to the bars relative to the tree and its tip labels. (Note: you can increase the canvas width in ``draw(width=...)`` to give more space to both). The relative lengths of bars will be scaled wtihin the depth region to represent differences in the data feature.

``width`` sets the width of bars between 0-1. The default of 0.8 leaves some room between bars, whereas 1.0 would place the bars flush against each other.


```python
# draw a tree
canvas, axes, m0 = tree.draw(layout="r", width=350)

# add bars mapped to tips on the tree
m1 = tree.annotate.add_tip_bars(axes=axes, data="Y", offset=0.0, depth=150, width=0.9);
```

The bars are plotted relative to each tip node on the drawn tree. Thus, if you are working with a tree that does not have aligned tips, you can use the drawing argument ``draw(tip_labels_align=True`` to align the tip positions. Alternatively, if you want the bars to be plotted next to unaligned tips, then you can do that too.

```python
# a tree with unaligned tips
tree2 = toytree.rtree.rtree(ntips=15, seed=123)

# draw a tree
canvas, axes, m0 = tree2.draw(layout="r", tip_labels_align=True, width=350)

# add bars mapped to tips on the tree
m1 = tree.annotate.add_tip_bars(axes=axes, data="Y");
```

## Data scale bar
Because bars typically represent real data, you will often want to add a scale bar to the barplot. This can be done using the [add_axes_scale_bar_to_mark](anno-add-axes-scale-bar-to-mark), which can be further customized. 

```python
# draw a tree
canvas, axes, m0 = tree.draw(layout="r", width=350)

# add bars mapped to tips on the tree
m1 = tree.annotate.add_tip_bars(axes=axes, data="Y")
tree.annotate.add_axes_scale_bar_to_mark(axes=axes, mark=m1, label="data (units)");
```

## Styling bars
### Style
The style dict can be used to set stroke and fill values that apply to all bars.

```python
c, a, m = tree.draw(width=350)
m1 = tree.annotate.add_tip_bars(
    a, "Y", style={"stroke": "black", "stroke-width": 1.0, "fill": "salmon"}
)
```

### Color and opacity
The ``color`` and ``opacity`` args can be used to set per-datum styles. These can be entered as a single value, a sequence of ntips values, as a str for a feature that is stored to the tree, or using the tuple-format to map a feature to a colormap. The example below maps the feature "Y" to the BlueRed colormap. 

```python
c, a, m = tree.draw(width=350)
m1 = tree.annotate.add_tip_bars(
    axes=a,
    data="Y",
    color=("Y", "BlueRed"),
    opacity=0.7,
    style={"stroke": "#262626", "stroke-width": 1.5, "stroke-opacity": 1.0},
)
```

## Circle/Annular bars

```python
c, a, m = tree.draw(width=550, layout="c")
m1 = tree.annotate.add_tip_bars(
    axes=a,
    data="Y",
    color=("Y", "BlueRed"),
    opacity=0.7,
    style={"stroke": "#262626", "stroke-width": 1.5, "stroke-opacity": 1.0},
    depth=80,
)
```

## Mask
The ``mask`` arg is used to subselect which tip nodes to show data for. The method ``ToyTree.get_tip_mask`` can be useful for creating a boolean mask by selecting tips by node idx or name.

```python
c, a, m = tree.draw(width=350)
m1 = tree.annotate.add_tip_bars(
    axes=a,
    data="Y",
    mask=tree.get_tip_mask(*range(3, 10)),
)
```
