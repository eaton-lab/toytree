# Annotate: add-edge-markers

The annotation method `add_edge_markers` is used to add markers to the midpoint of an edge. 

Using `add_edge_markers` and related `add_edge_[...]` methods requires a little more thought than other annotation methods. This is because feature data on edges can be interpreted in many different ways. For example, it would make sense to plot the 'dist' value of each edge on a tree, but it makes less sense to plot the 'support' values on every edge, since only internal edges of the tree truly have supports. Thus, the `mask` argument plays an important role with this tool to select which edges you want to plot data onto, depending on the type of data you intend to show.

```python
import toytree

# an example tree with continuous (X) and categorical (Y) data
tree = toytree.rtree.bdtree(ntips=10, seed=123)
tree.pcm.simulate_continuous_trait(name="X", inplace=True)
tree.pcm.simulate_discrete_trait(nstates=3, trait_name="Y", inplace=True);
```

```python
canvas, axes, m0 = tree.draw()
m1 = tree.annotate.add_edge_markers(axes, marker="s", color="red", size=10, xshift=-5)
m2 = tree.annotate.add_edge_markers(axes, marker="s", color="blue", size=10, xshift=5)
m2 = tree.annotate.add_edge_markers(
    axes, marker="s", color=("X", "BlueRed"), size=10, xshift=5
)
m2 = tree.annotate.add_edge_markers(
    axes, marker="s", color=("Y", "Set2"), size=10, xshift=-5
);
```

```python
canvas, axes, m0 = tree.draw()
m1 = tree.annotate.add_edge_markers(
    axes, marker="s", size=15, color="red", mask=tree.get_node_mask(14)
)
```

## Selected edges (mask)
It is often desirable to be able to select which edge to add markers to. This is best accomplished using the `mask` argument, especially in combination with the tree objects `.get_node_mask()` method. 

### mask: bool
Setting `mask=False` will add a marker to all edges in the tree, i.e., no edges are masked.

```python
canvas, axes, m0 = tree.draw()
m1 = tree.annotate.add_edge_markers(axes, mask=False);
```

### mask: get_node_mask

```python
canvas, axes, m0 = tree.draw()
m1 = tree.annotate.add_edge_markers(axes, mask=tree.get_node_mask(11, 12, 13));
```

```python
canvas, axes, m0 = tree.draw()
m1 = tree.annotate.add_edge_markers(
    axes,
    mask=tree.get_node_mask(
        11,
    ),
);  # BUG
```
