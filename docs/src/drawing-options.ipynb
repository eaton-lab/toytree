# Drawing options

`ToyTree.draw()` is the main entry point for tree figures in `toytree`. This page focuses on the arguments that change the geometry and styling of one tree drawing.

If you are new to the drawing API, start with [drawing-basics.md](drawing-basics.md) for the returned `canvas`, `axes`, and `mark` objects. If you want to attach features before mapping them to colors, sizes, or labels, see [core-data.md](core-data.md).

Most style arguments follow one of three patterns:
- a single value applies everywhere;
- a sequence applies values in tip or node order;
- some size and color arguments can map a tree feature into a visual scale.


## Example tree

We will use one small local tree throughout this page. It includes a few extra features so the later examples can show label, color, and size mappings without fetching outside data.


````python
import numpy as np

import toytree

tree = toytree.rtree.bdtree(ntips=8, seed=123)

support_map = {
    idx: value
    for idx, value in zip(range(tree.ntips, tree.nnodes), [100, 96, 91, 84, 77, 71, 65])
}
group_map = {
    idx: ("Test" if idx % 2 == 0 else "Reference") for idx in range(tree.nnodes)
}
score_map = {idx: round(0.5 + idx / 12, 2) for idx in range(tree.nnodes)}

tree = tree.set_node_data("support", support_map, default=np.nan)
tree = tree.set_node_data("group", group_map, default="Reference")
tree = tree.set_node_data("score", score_map, default=np.nan)
````

## Quick start

A good way to think about `draw()` is that `tree_style` chooses a starting look, and the other arguments adjust specific parts of that look. The shorter alias `ts` is convenient when you are experimenting in a notebook. Built-in styles include `n`, `s`, `p`, `o`, `c`, `d`, `b`, `u`, and `r`.

`width`, `height`, and `label` change the figure frame rather than the tree data. If you pass an existing `axes`, the tree is drawn onto that axes and the size arguments are ignored because the axes already exist.


````python
tree.draw(ts="s", width=420, height=300, label="Styled example");
````

## Layout

`layout` controls the overall direction of the drawing. The most common values are `"r"`, `"l"`, `"u"`, and `"d"` for right, left, up, and down. Circular trees use `"c"` or an arc such as `"c90-270"`. Strings beginning with `"un"` switch to an unrooted layout.

Two other arguments often matter with layout:
- `use_edge_lengths=False` ignores branch-length scaling and aligns the tips.
- `edge_type` changes the branch path shape: `"p"` is the default phylogram path, `"c"` is a straight cladogram segment, and `"b"` uses a Bezier curve.

Add `scale_bar=True` when branch lengths represent a real distance or time scale. If you pass a number such as `scale_bar=1e6`, the displayed tick labels are divided by that value.


````python
tree.draw(layout="d", ts="s");
````

Circular layout is often used for very large trees. With so many tips you typically cannot see tip labels and so it is better to suppress them.

````python
# circular layout
big_tree = toytree.rtree.bdtree(ntips=500)
big_tree.draw(layout="c", tip_labels=False);
````

A partially circular (fan) layout can be designated as ``cN`` or ``cX-Y`` setting the starting or starting end end polar coordinates in degrees. 

````python
# fan layout
big_tree = toytree.rtree.bdtree(ntips=50)
big_tree.draw(layout="c0-180", tip_labels=True);
````

````python
tree.draw(layout="r", edge_type="c", use_edge_lengths=False, ts="o");
````

### Unit multiplier
If you set a numeric value for ``scale_bar`` it will not only add the scale bar but also divide the units by the input value. This can be used to easily scale the units to more readable values when they are in units of millions of years, or expected substitutions per site.

````python
tree.draw(scale_bar=1000);
````

## Tip labels

`tip_labels=True` uses the tip `name` feature. You can hide the labels, replace them with another tip feature, or pass a full sequence of strings in tip order. `tip_labels_align=True` adds guide lines between the tip positions and the label text, which is especially useful when labels have different widths.

`tip_labels_colors` accepts a single color, a per-tip sequence, or a feature-to-colormap tuple such as `("score", "BlueRed")`. `tip_labels_style` is the place to change shared text styling such as font size or weight. `tip_labels_angles` is mainly useful on circular layouts when you want to override the automatic label rotation.

Color values follow [Toyplot color syntax](https://toyplot.readthedocs.io/en/stable/colors.html#color-values).


````python
custom_tip_labels = [
    f"{name} ({group})"
    for name, group in zip(tree.get_tip_labels(), tree.get_tip_data("group").values)
]
tip_colors = [
    "#3a6ea5" if group == "Test" else "#dd8452"
    for group in tree.get_tip_data("group").values
]

tree.draw(
    tip_labels=custom_tip_labels,
    tip_labels_align=True,
    tip_labels_colors=tip_colors,
    tip_labels_style={"font-size": 13, "font-weight": 600},
);
````

## Nodes

Node arguments are useful when you want to show support values, ancestral states, or any other feature stored on the tree. `node_mask` controls which nodes are drawn. A 3-item tuple is often the clearest form: `(show_tips, show_internal, show_root)`.

`node_labels` can show `"idx"`, another feature name such as `"support"`, or a formatting tuple like `("score", "{:.2f}")`. `node_sizes` and `node_colors` accept a single value, a per-node sequence, or a feature mapping. For example, `("support", 8, 18)` maps support values into a size range, while `("support", "Spectral")` maps them into colors.

`node_markers` changes the marker shape. `node_hover` adds tooltip content for interactive notebook views. `node_style` holds shared marker styling such as stroke or fill opacity, and per-node arguments override the matching shared style fields when both are set. Marker names follow [Toyplot marker syntax](https://toyplot.readthedocs.io/en/stable/markers.html).


````python
tree.draw(
    node_mask=(False, True, True),
    node_labels="support",
    node_sizes=("support", 8, 18),
    node_colors=("support", "Spectral"),
    node_style={"stroke": "#333333", "stroke-width": 1.5},
);
````

````python
tree.draw(
    node_mask=(False, True, False),
    node_labels=("score", "{:.2f}"),
    node_sizes=14,
    node_markers="s",
    node_as_edge_data=True,
);
````

### node_sizes
In the default tree style node sizes are set to zero, meaning that they will not be shown. Node sizes are described in pixel units as floats or ints, which scale node marker sizes. A single value can be entered to apply to all nodes, or a sequence of values of length nnodes can be entered to scale nodes to different sizes. In this case, the values should be enetered in node idxorder, which is best done by extracting data from the tree object itself, as described in [Data/Features](/toytree/data). See also [Range-Mapping](/toytree/range-mapping) for examples of scaling node sizes to data features. 

## Edges

Each drawn edge is associated with its child node, so per-edge sequences follow node order. `edge_colors` and `edge_widths` use the same single-value, sequence, or feature-mapping patterns as node colors and sizes.

`edge_style` changes the main branch styling. `edge_align_style` is more specific: it styles the extra alignment lines added by `tip_labels_align=True`. This is a good place to set a lighter color or dashed pattern so the aligned guide lines stay visually separate from the real branches.


````python
tree.draw(
    edge_colors=("support", "BlueRed"),
    edge_widths=("score", 1.5, 5),
    edge_style={"stroke-opacity": 0.85},
    tip_labels_align=True,
    edge_align_style={"stroke": "#bbbbbb", "stroke-dasharray": "4,2"},
);
````

## Positioning

These arguments matter most when you are fitting a tree into a larger figure or working with long labels.

- `padding` changes the whitespace inside the axes.
- `xbaseline` and `ybaseline` shift the tree within the current coordinate system.
- `shrink` reserves more room in the tip-label direction so the tree body does not crowd the labels.
- `axes` lets you draw onto an existing Toyplot Cartesian instead of creating a new one.

A few typical patterns look like this:

```python
tree.draw(width=520, height=300, padding=30, label="Example tree")

tree.draw(tip_labels_align=True, shrink=30)

canvas, axes, mark = tree.draw()
other_tree.draw(axes=axes, layout="l")
```


## Advanced layouts

`fixed_order` and `fixed_position` are useful when the natural traversal order is not the order you want to show on the tip axis. This is common when comparing discordant trees, aligning several trees to the same tip order, or leaving visible gaps for missing or extinct taxa.

`interior_algorithm` changes how internal nodes are placed along that fixed tip axis on linear layouts. The default `0` uses the midpoint of immediate children. Other values use descendant-tip summaries and are most noticeable when tips are unevenly spaced.

`admixture_edges` adds one or more reticulation arrows on top of the tree. Use it when you already have admixture events to overlay on an otherwise tree-like drawing.


````python
reversed_order = list(reversed(tree.get_tip_labels()))
spaced_positions = [0, 1, 2, 4, 6, 7, 9, 10]

tree.draw(
    layout="d",
    fixed_order=reversed_order,
    fixed_position=spaced_positions,
    interior_algorithm=3,
    ts="s",
);
````

A compact `admixture_edges` example looks like this:

````python
tree.draw(ts="c", admixture_edges=[(6, 4)]);
````

## Common patterns

Here are a few combinations that cover many day-to-day drawing tasks.

````python
# A clean rooted tree with a scale bar.
tree.draw(ts="s", layout="r", scale_bar=True)

# Show support values on the branches instead of at the nodes.
tree.draw(
    node_mask=(False, True, False),
    node_labels="support",
    node_sizes=14,
    node_as_edge_data=True,
)

# Use a fixed tip order when comparing several related trees.
order = list(reversed(tree.get_tip_labels()))
tree.draw(layout="d", fixed_order=order, ts="o");
````
