# Add scale bars to tree or annotation marks

```python
import toytree

tree = toytree.rtree.bdtree(ntips=12, seed=123)
tree.pcm.simulate_continuous_trait(name="X", root_state=5, inplace=True, seed=123);
```

```python
# draw a tree and annotation mark with data
c, a, m = tree.draw(width=400)
m1 = tree.annotate.add_tip_bars(a, data="X", depth=150)

# add scale bar to trait bars data axis
tree.annotate.add_axes_scale_bar_to_mark(
    a,
    m1,
    label="trait (units)",
);
```

## Example

```python
# draw a tree and annotation mark with data
c, a, m = tree.draw(width=400)
m1 = tree.annotate.add_tip_bars(a, data="X")

# add scale bar to tree depth data axis
tree.annotate.add_axes_scale_bar_to_tree(a, label="depth (units)")

# add scale bar to trait bars data axis
tree.annotate.add_axes_scale_bar_to_mark(a, m1, label="trait (units)");
```
