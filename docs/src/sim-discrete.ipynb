# Simulate discrete (Markov) states

To simulate transitions between discrete characters along the edges of a phylogeny we can use a continuous-time Markov chain (CTMC). This models the transitions between n possible discrete states as a function of time (edge lengths). In a CTMC a **rate matrix (Q)** is used to describe rates of transition per unit time, where the probability of transitioninng between states in a small step of time (dt) is given by an exponentiation of this matrix $P(t)=e^{Qt}$.

### Relative rates matrix
A relative rates matrix R specifies the relative transition rates between states, but it does not have absolute scaling or satisfy the Markov constraint (rows summing to 0). It looks like below, where $r_{01}$ represents the relative rate of transitioning from state 0 to state 1.

$$
R = \begin{bmatrix}  
0 & r_{01} \\  
r_{10} & 0 \\  
\end{bmatrix}
$$

```python
mm = toytree.pcm.get_markov_model(nstates=2, relative_rates=[[0, 5], [5, 0]])
mm.relative_rates
```

### Q-matrix
To initiate a Markov model we must define the state space in terms of the number (and optionally names) of the possible discrete states, as well as the transition probability matrix. For a 2-state model the transition matrix should be 2 x 2 in shape. 

For a continuous-time Markov model with ($n$) states, the **Q matrix** ($Q$) is an ($n \times n$) matrix where:

- ($Q_{ij}$) (**off-diagonal elements**) represents the rate at which transitions occur from state \( i \) to state ($j$) ($i \neq j$)).
- ($Q_{ii}$) (**diagonal elements**) is set so that each row sums to **0**, ensuring the model maintains a probability distribution.


For a 2-state continuous-time Markov model, the Q matrix looks like:

$$
Q = \begin{bmatrix}  
- \lambda & \lambda \\  
\mu & -\mu  
\end{bmatrix}
$$

where:
- ($\lambda$) is the transition rate from state **0 → 1**.
- ($\mu$) is the transition rate from state **1 → 0**.
- The diagonal entries ensure each row sums to 0:  
$$
  Q_{ii} = - \sum_{j \neq i} Q_{ij}
$$
