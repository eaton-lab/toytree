# Add tip tiles

The annotation ``add_tip_tiles`` adds a tile (rectangle or annular wedge) across the tips of a tree. It can be used to display groupings (e.g., taxonomic clades) or trait information by setting different colors. See also [add_tip_bars](anno-add_tip_bars).

```python
import toytree

# an example tree
tree = toytree.rtree.bdtree(ntips=15, seed=123)

# simulate a categorical and quantitative trait
tree.pcm.simulate_discrete_trait(nstates=2, trait_name="X", inplace=True, seed=123)
tree.pcm.simulate_continuous_trait(name="Y", inplace=True, seed=321);
```

## Example
The tip tile markers are automatically fit over the width of tip labels. This makes tiles convenient for highlighting groups of tips that form a clade, or share a character state.

In this example I select the binary trait "X" and map it to the colors blue and red.

```python
c, a, m = tree.draw()
tree.annotate.add_tip_tiles(a, color=("X", ["blue", "red"]), opacity=0.5);
```

## Default style
The default style applies a light grey color to tiles that are placed behind the tips, spanning from depth position 0 (at the tip) to 15px past the farthest tip. 

```python
c, a, m = tree.draw()
tree.annotate.add_tip_tiles(a);
```

## Tile offset and depth

You can set the ``offset`` and ``depth`` arguments to control the distance from the tip nodes that tiles start, and the depth that they will span. The default ``offset`` is 0px, and default ``depth`` is None, which automatically extends tiles over the tips.

Below, we manually set the ``offset`` and ``depth`` to numeric values to place tiles past the tip labels.

```python
c, a, m = tree.draw()
tree.annotate.add_tip_tiles(a, offset=45, depth=25);
```

The positioning arguments specify distances, and are thus always positive, regardless of the tree layout.

```python
c, a, m = tree.draw(layout="d")
tree.annotate.add_tip_tiles(a, offset=45, depth=25);
```

Multiple tip tile annotations can be placed on the same drawing. This provides one simple way to make a grid.

```python
c, a, m = tree.draw(layout="d")
tree.annotate.add_tip_tiles(a, offset=45, depth=25, color=("X", "Set2"))
tree.annotate.add_tip_tiles(a, offset=70, depth=25, color=("Y", "BlueRed"));
```

## Styling tiles

The ``color`` argument can be passed as a single color, a list of colors equal to the number of tiles, as a string name of a feature on the tree, or as a (feature, colormap) tuple to map a feature to a colormap.

In addition, you can set the ``opacity`` and use the ``style`` dict to set CSS styles including stroke, stroke-width and others.

```python
c, a, m = tree.draw(layout="d")
tree.annotate.add_tip_tiles(
    axes=a,
    offset=45,
    depth=25,
    color=("idx", "Spectral"),
    opacity=0.8,
    style={"stroke": "#262626", "stroke-width": 2},
);
```

## Circle/Annular tiles
Annular tiles follow the same logic, but are shaped to follow polar rather than cartesian coordinates.

```python
# simulate a big tree with a discrete trait
big_tree = toytree.rtree.bdtree(200)
big_tree.pcm.simulate_discrete_trait(nstates=2, trait_name="X", inplace=True)

# draw the tree in layout 'c' and add tiles for trait 'X'
c, a, m = big_tree.draw(layout="c", width=450, tip_labels=False)
big_tree.annotate.add_tip_tiles(
    a, offset=5, depth=25, opacity=0.5, color=("X", ["salmon", "darkcyan"])
);
```

This can similarly apply to "fan" style tree layouts.

```python
c, a, m = tree.draw(layout="c90", width=450)
tree.annotate.add_tip_tiles(a, offset=5, depth=35, opacity=0.2, color="grey")
tree.annotate.add_tip_tiles(
    a, offset=40, depth=35, color=("Y", "BlueRed"), style={"stroke": "#262626"}
);
```

## Mask
You can use the ``mask`` argument to subselect which tips to show tiles for. As with other annotation methods, there are several shortcuts for selecting nodes. Here we use the ``get_tip_mask`` method.

```python
c, a, m = tree.draw(layout="d")
tree.annotate.add_tip_tiles(
    axes=a,
    offset=45,
    depth=25,
    style={"stroke": "#262626", "stroke-width": 2},
    mask=tree.get_tip_mask("r0", "r1", "r2", "r8", "r11"),
);
```

## Label a tile row/column
I suggest using the very flexible `Cartesian.text()` method when you want to add text labels to annotation Marks. 

This takes input like ``(X, Y, 'text', style={...})``. The X and Y coordinates must reference a position in data coordinates. You will typically want to select a tip node, such as idx=0 at (0, 0), or idx=ntips at (ntips, 0). From there, you can use the style arguments to position the label by nudging it in pixel units.

In this example, we created a row of tiles at 40 and 70 px units from the tips. So I shifted the labels down by 50 and 80 units to align with the center of the tile rows.

```python
c, a, m = tree.draw(layout="d", scale_bar=True)

m1 = tree.annotate.add_tip_tiles(
    axes=a,
    offset=40,
    depth=20,
    color="X",
    style={"stroke": "#262626", "stroke-width": 2},
)
m2 = tree.annotate.add_tip_tiles(
    axes=a,
    offset=70,
    depth=20,
    color="Y",
    style={"stroke": "#262626", "stroke-width": 2},
)

# add a label to a tile row/column: positioned by px units
kwargs = {"fill": "black", "text-anchor": "end"}
a.text(
    0,
    0,
    "<b>Trait X</b>",
    style={"baseline-shift": -50, "-toyplot-anchor-shift": -15, **kwargs},
)
a.text(
    0,
    0,
    "<b>Trait Y</b>",
    style={"baseline-shift": -80, "-toyplot-anchor-shift": -15, **kwargs},
);
```
